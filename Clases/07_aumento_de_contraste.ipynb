Aumento de Contraste de una imagen
-el contreaste es una funcion
-el contrate multiplica al pixel 
Aumento de brillo y luminosidad 
-El brillo aumenta la cantida de luz percibida en una imagen

Contraste

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def contraste(img, alfa = 1.0, beta = 0):
    img_nueva = np.clip(alfa * img + beta, 0, 255 ).astype(np.uint8)
    return img_nueva
```

```python
img = cv2.imread('..//img/noche_estrellas.jpg', cv2.IMREAD_GRAYSCALE)

img_contraste = contraste(img, 1.7, 0)

plt.rcParams['figure.figsize'] = [10, 5]

plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(img, cmap='gray')

plt.subplot(1,2,2)
plt.title('Contraste')
plt.imshow(img_contraste, cmap='gray')

plt.show()
```

Contraste con Canales(R,G,B)

```python
def ContrasteColor(img, alpha = 1, beta = 0):
    img_nueva = cv2.convertScaleAbs(img, alpha = alpha, beta = beta)
    return img_nueva
```

```python
img = cv2.imread('..//img/noche_estrellas.jpg', cv2.IMREAD_COLOR)
red, green, blue = cv2.split(img)

contraste_blue = ContrasteColor(blue, alpha = 0, beta = 0)
contraste_red = ContrasteColor(red, alpha = 0, beta = 0)
contraste_green = ContrasteColor(green, alpha = 1.2, beta = 1)

finImg = cv2.merge((contraste_red, contraste_green, contraste_blue))


plt.figure(figsize = (20, 10))

plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1,2,2)
plt.title('Contraste')
plt.imshow(finImg)
plt.axis('off')

plt.show()
```
