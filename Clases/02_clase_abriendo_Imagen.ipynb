```python
from PIL import Image
from matplotlib import pyplot as plt
from collections import Counter
from scipy import ndimage as ndi
from skimage import feature
import numpy as np
import cv2

# Cargar una imagen con PIL
image_path = r"..//img\imgpru.jpg"
original_image = Image.open(image_path)

# Mostrar la imagen original
plt.imshow(original_image)
plt.title('Imagen Original')
plt.show()

# Convertir la imagen a un array de numpy para procesamiento
image_array = np.array(original_image)

# Operaciones de procesamiento de imagen con OpenCV
gray_image = cv2.cvtColor(image_array, cv2.COLOR_RGB2GRAY)
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)

# Mostrar la imagen procesada con OpenCV
plt.imshow(blurred_image, cmap='gray')
plt.title('Imagen Procesada con OpenCV')
plt.show()

# Operaciones de procesamiento de imagen con scikit-image
edges = feature.canny(blurred_image, sigma=1)

# Mostrar los bordes detectados con scikit-image
plt.imshow(edges, cmap='gray')
plt.title('Bordes Detectados con scikit-image')
plt.show()

# Operaciones de procesamiento de imagen con scipy
labeled_image, num_labels = ndi.label(edges)
label_counts = Counter(labeled_image.flatten())

# Mostrar la imagen etiquetada y el número de regiones encontradas
plt.imshow(labeled_image, cmap='nipy_spectral')
plt.title(f'Imagen Etiquetada - {num_labels} Regiones')
plt.show()

# Estadísticas sobre las regiones encontradas
region_sizes = list(label_counts.values())
mean_size = statistics.mean(region_sizes)
median_size = statistics.median(region_sizes)

print(f"Número total de regiones encontradas: {num_labels}")
print(f"Tamaño medio de las regiones: {mean_size}")
print(f"Tamaño mediano de las regiones: {median_size}")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


image = cv2.imread(r'..//img/yop.jpg', cv2.IMREAD_GRAYSCALE)
# guardamos profundidad para mostrar
original_img = image.dtype
print('profundidad original de color: ', original_img)
# transformacion de imagen a 16 bits
image_16bits = np.uint16(image)
nueva_img = image_16bits.dtype
print('profundidad nueva de color: ', nueva_img)
# guardamos la imagen tratada
cv2.imwrite('..//img/imagen16bits.jpg', image_16bits)

img_guardada = cv2.imread('..//img/imagen16bits.jpg', cv2.IMREAD_UNCHANGED)

plt.figure(figsize=(10,5))
plt.subplot(121)
plt.title('Mi imagen bonita')
plt.imshow(image, cmap='gray')

plt.subplot(122)
plt.title('Mi imagen bonita tratada 16 bits')
plt.imshow(img_guardada, cmap='gray')

plt.show()
```
