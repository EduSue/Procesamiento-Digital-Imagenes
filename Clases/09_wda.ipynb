```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import imutils
```

```python
img =  cv.imread('..//img/chica.jpg', 0)

t1 = np.fft.fft2(img)
t2 = np.fft.fftshift(t1)
fila, columna = img.shape
fila1, columna1 = fila // 2, columna // 2

t2[fila - 5: fila + 5, columna1 - 5: columna1 + 5] = 0
t3 = np.fft.ifftshift(t2)

imgFin = np.fft.irfft2(t3)
imgFin = np.abs(imgFin)



espectro = 20 * np.log(np.abs(t2))

plt.figure(figsize=(10, 10))

plt.subplot(1,3,1)
plt.title('Imagen Original')
plt.imshow(img, cmap='gray')

plt.subplot(1,3,2)
plt.title('Imagen Transformada')
plt.imshow(imgFin, cmap='gray')

plt.subplot(1,3,3)
plt.title('Imagen Espectro')
plt.imshow(espectro, cmap='gray')

plt.tight_layout()
plt.show()
```

```python
img = cv.imread('..//img/chica.jpg')

escala = 0.6

img_redim = cv.resize(img, None, fx=escala, fy=escala)

plt.figure(figsize=(10, 10))

plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))

plt.subplot(1,2,2)
plt.title('Imagen escalada')
plt.imshow(cv.cvtColor(img_redim, cv.COLOR_BGR2RGB))

plt.tight_layout()
plt.show()
```

```python
img = cv.imread('..//img/chica.jpg')

x = 120
y = 60

matriz_tras = np.float32([[1, 0, x], [0, 1, y]])

img_tras = cv.warpAffine(img, matriz_tras, (img.shape[1], img.shape[0]))

plt.figure(figsize=(10, 5))

plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(img)

plt.subplot(1,2,2)
plt.title('Imagen trasladada')
plt.imshow(img_tras)

plt.tight_layout()
plt.show()
```
