```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen_aliasing = cv2.imread('..//img/imgali.png')

imagen_gris = cv2.cvtColor(imagen_aliasing, cv2.COLOR_BGR2GRAY)

imagen_antialiasing = cv2.GaussianBlur(imagen_gris, (5, 5), 0)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(imagen_gris, cmap='gray')
plt.title('Imagen con Aliasing')

plt.subplot(1, 2, 2)
plt.imshow(imagen_antialiasing, cmap='gray')
plt.title('Imagen sin Aliasing')

plt.show()
```

```python
# Moire
img1 = cv2.imread('..//img/cabeimg.png')
img1Fix = cv2.medianBlur(img1,7)
fig = plt.figure(num=None, figsize=(80, 80), dpi=80, facecolor='w', edgecolor='k')
fig1 = fig.add_subplot(10, 10, 1)
fig2 = fig.add_subplot(10, 10, 2)
fig1.title.set_text('Orignal image')
fig1.imshow(img1)
fig2.title.set_text('Mascara 7x7 aplicada')
fig2.imshow(img1Fix)
plt.show()
```
