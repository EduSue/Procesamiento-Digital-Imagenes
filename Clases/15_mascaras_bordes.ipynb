```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('..//img/e3e.jpg', 0)

mascaras = [
    np.array([[1, -1]]),  
    np.array([[1], [-1]]),  
    np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]]),
    np.array([[1,1,1], [0,0,0], [-1,-1,-1]]),  
    np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]]),
    np.array([[1,2,1], [0,0,0], [-1, -2, -1]])
]

plt.figure(figsize=(15, 10))

plt.subplot(4, 3, 1)
plt.title('Original')
plt.imshow(img, cmap='gray')
plt.axis('off')

for i, mascara in enumerate(mascaras):
    img_conv = cv2.filter2D(img, -1, mascara)
    plt.subplot(4, 3, i+2)
    plt.title(f'Mscara {i+1}')
    plt.imshow(img_conv, cmap='gray')
    plt.axis('off')

img_conv = cv2.filter2D(img, -1, mascaras[0])
img_conv1 = cv2.filter2D(img_conv, -1, mascara[1])

plt.subplot(4, 3, 8)
plt.title('Convinacion 1')
plt.imshow(img_conv1, cmap='gray')
plt.axis('off')

img_conv = cv2.filter2D(img, -1, mascaras[2])
img_conv2 = cv2.filter2D(img_conv, -1, mascaras[3])

plt.subplot(4, 3, 9)
plt.title('Convinacion 2')
plt.imshow(img_conv2, cmap='gray')
plt.axis('off')

img_conv = cv2.filter2D(img, -1, mascaras[4])
img_conv3 = cv2.filter2D(img_conv, -1, mascaras[5])

plt.subplot(4, 3, 10)
plt.title('Convinacion 3')
plt.imshow(img_conv3, cmap='gray')
plt.axis('off')

t_lower = 50  
t_upper = 150  
aperture_size = 5 
  
edge = cv2.Canny(img, t_lower, t_upper) 
  
plt.subplot(4, 3, 11)
plt.title('Canny')
plt.imshow(edge, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
