```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('..//img/img1e.jpg', 0)


mascaras = [
    np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]]),  
    np.array([[0, 1, 0], [1, 0, 1], [0, -1, 0]]),  
    np.array([[1, -1, 1], [-1, 9, -1], [1, -1, 1]]),
    np.array([[0, -1, 0], [1, -8, 1], [0, -1, 0]]),  
    np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]]),
    np.array([[-1, -1, -1], [-1, 16, -1], [-1, -1, -1]]),
    np.array([[1, 0, -1], [0, 3, 0], [-1, 0, 1]]),  
    np.array([[1, -1, 1], [-1, 3, -1], [1, -1, 1]]),  
    np.array([[0, 1, 0], [-1, 5, -1], [0, 1, 0]]),
    np.array([[1, -2, 1], [-2, 8, -2], [1, -2, 1]]), 
    np.array([[-1, 2, -1], [2, -4, 2], [-1, 2, -1]]) 
]

plt.figure(figsize=(12, 9))

plt.subplot(4, 3, 1)
plt.title('Original')
plt.imshow(img, cmap='gray')
plt.axis('off')

for i, mascara in enumerate(mascaras):
    img_conv = cv2.filter2D(img, -1, mascara)
    plt.subplot(4, 3, i+2)
    plt.title(f'Mscara {i+1}')
    plt.imshow(img_conv, cmap='gray')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
