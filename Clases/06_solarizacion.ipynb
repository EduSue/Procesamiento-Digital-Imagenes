```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


def solarizar (imagen, umbral):
    imgSolarizada = np.where(imagen < umbral, imagen, 255 - imagen)
    return imgSolarizada

def posterizar(imagen, nivel):
    f_nvl = 255/ nivel
    img_poster = np.round(imagen/f_nvl) * f_nvl
    return img_poster.astype(np.uint8)

imagenRecibida = cv2.imread('..//img/cancer.jpg')
img_gris = cv2.cvtColor(imagenRecibida, cv2.COLOR_BGR2GRAY)
img_original = cv2.cvtColor(imagenRecibida, cv2.COLOR_BGR2RGB)
umbral = 100
nivel = 4

img_sol = solarizar(img_gris, umbral)
img_post = posterizar(img_gris, nivel)

plt.figure(figsize=(15, 7))

plt.subplot(1, 3, 1)
plt.imshow(img_original, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img_sol, cmap='gray')
plt.title('Solarizado')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(img_post, cmap='gray')
plt.title('Posterizado')
plt.axis('off')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [10,5]
tamanio_img = 100

#imagen negra cuando en todo ponga 255
R = np.ones((tamanio_img, tamanio_img)) * 255
G = np.ones((tamanio_img, tamanio_img)) * 0
B = np.ones((tamanio_img, tamanio_img)) * 0

imagen = np.dstack((R,G,B))
imagen = np.array(imagen, dtype=int)

plt.figure(figsize=(15, 7))  
plt.imshow(imagen, cmap='gray')
plt.title('Rojo')
plt.axis('off')
```
