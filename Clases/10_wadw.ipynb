```python
# Filtro mascara de figuras

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('..//img/chica.jpg', cv.IMREAD_GRAYSCALE)

alto, ancho = img.shape[:2]

mascara = np.zeros_like(img)
circulo = (ancho // 2, alto // 3)
# circulo = (ancho // 2, alto // 2)
radio = 180
# radio = 100

cv.circle(mascara, circulo, radio, (255), -1)

img_fin = cv.bitwise_and(img, img, mask=mascara)

plt.figure(figsize=(10, 5))

plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(img, cmap='gray')

plt.subplot(1,2,2)
plt.title('Imagen mascara')
plt.imshow(img_fin, cmap='gray')

plt.tight_layout()
plt.show()

```

```python


mascara = np.zeros_like(img)
circulo = (ancho // 2, alto // 3)
# circulo = (ancho // 2, alto // 2)
radio = 180
# radio = 100

cv.circle(mascara, circulo, radio, (255), -1)

img_fin = img * (mascara / 255) + (img // 2) * (1 - mascara / 255)

plt.figure(figsize=(10, 5))

plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(img, cmap='gray')

plt.subplot(1,2,2)
plt.title('Imagen mascara')
plt.imshow(img_fin, cmap='gray')

plt.tight_layout()
plt.show()
```

```python
img = cv.imread('..//img/imagen1.jpg')
i_g = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

mask_deg = np.linspace(1, 0, i_g.shape[0])[:, np.newaxis] # mascara degradado
mask_fin = np.tile(mask_deg, (1, i_g.shape[1]))
img_fin = (i_g.astype(np.float32) * mask_fin).astype(np.uint8)

plt.imshow(img_fin, cmap='gray')
```
