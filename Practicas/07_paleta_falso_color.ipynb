```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
vecsin = [
    "nada",
    cv2.COLORMAP_AUTUMN, 
    cv2.COLORMAP_BONE,
    cv2.COLORMAP_JET,
    cv2.COLORMAP_WINTER,
    cv2.COLORMAP_RAINBOW,
    cv2.COLORMAP_OCEAN,
    cv2.COLORMAP_SUMMER,
    cv2.COLORMAP_SPRING,
    cv2.COLORMAP_COOL,
    cv2.COLORMAP_HSV,
    cv2.COLORMAP_PINK,
    cv2.COLORMAP_HOT,
    cv2.COLORMAP_PARULA,
    cv2.COLORMAP_MAGMA,
    cv2.COLORMAP_INFERNO,
    cv2.COLORMAP_PLASMA,
    cv2.COLORMAP_VIRIDIS,
    cv2.COLORMAP_CIVIDIS,
    cv2.COLORMAP_TWILIGHT,
    cv2.COLORMAP_TWILIGHT_SHIFTED,
    cv2.COLORMAP_TURBO,
    cv2.COLORMAP_DEEPGREEN
    ]
  
```

```python
routita = '..//img/imgpru.jpg'
img = cv2.imread(routita, 1)
plt.rcParams['figure.figsize'] = [10, 10]
for i in range(len(vecsin)):
    numerin = i + 1
    if vecsin[i] != "nada":
        imC = cv2.applyColorMap(img, vecsin[i])
        plt.subplot(6, 4, numerin)
        plt.imshow(imC)
    else:
        plt.subplot(6, 4, numerin)
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title("Imagen " + str(numerin))

plt.tight_layout()
plt.show()
```
