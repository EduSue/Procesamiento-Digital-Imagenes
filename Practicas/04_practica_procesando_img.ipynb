```python
def mostarImg(imagen, tittle, subp = 111, cvc = cv2.IMREAD_COLOR, cm = 'nada'):
    route = (imagen)
    img = cv2.imread(route, cvc)
    plt.subplot(subp)
    plt.imshow(img)
    plt.title(tittle)
    if(cm != 'nada'):
        plt.imshow(img, cmap=cm)
    else:
        plt.imshow(img)
    plt.axis('off')
```
