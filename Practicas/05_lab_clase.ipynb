```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Umbralizacon binaria y binaria invetida

imgB = cv2.imread('..//img/noche_estrellas.jpg', 0)
# aca estoy recogiendo la imagen y colocandola con umbralizacion binaria y ek guion bajo es una forama de declarar en python para una variable que no usaremos
# en el guion bajo va el umbral 
_, umbralB = cv2.threshold(imgB, 127, 255, cv2.THRESH_BINARY)
# Umbralizacion binaria inversa 
_, umbralBinv = cv2.threshold(imgB, 127, 255, cv2.THRESH_BINARY_INV)

umb_adap = cv2.adaptiveThreshold(imgB, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)

umb_adap_gaus = cv2.adaptiveThreshold(imgB, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)


def histograma(image):
    histo = cv2.calcHist([image], [0], None, [256], [0, 256])

    plt.figure(figsize = (10, 5))
    plt.plot(histo, color = 'purple')
    plt.title('Histograma')
    plt.xlabel('intensidad_pixel')
    plt.ylabel('frecuencia_pixel')
    # plt.show()




plt.figure(figsize = (20, 10))

plt.subplot(2,3,1)
plt.title('Imagen original')
# plt.imshow(cv2.cvtColor(imgB, cv2.COLOR_BGR2RGB))
plt.imshow(imgB, cmap='gray')
# plt.axis('off')


plt.subplot(2,3,2)
plt.title('Imagen binarizada')
plt.imshow(umbralB, cmap='gray')
# plt.axis('off')

plt.subplot(2,3,3)
plt.title('Imagen binarizada inversa')
plt.imshow(umbralBinv, cmap='gray')
# plt.axis('off')

plt.subplot(2,3,4)
plt.title('Umbralizacion adaptativa Mean')
plt.imshow(umb_adap, cmap='gray')
# plt.axis('off')

plt.subplot(2,3,5)
plt.title('Imagen adaptativa Gauss')
plt.imshow(umb_adap_gaus, cmap='gray')
# plt.axis('off')

histograma(image = imgB)

plt.show()
```
